```python
from numpy import median
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df = df.rename(columns={"purchase_amount_(usd)" : "purchase_amount"})
```

```python
df.columns
```

```python
# Create a columns age_group
labels = ["Young Adult", "Adult","Middle-aged","Senior"]
df["age_group"] = pd.qcut(df["age"],q=4, labels = labels)
```

```python
df[["age","age_group"]].head(10)
```

```python
#Create columns purchase_frequency_days
frequency_mapping = {
    "Fortnightly":14,
    "Weekly":7,
    "Monthly":30,
    "Quaterly":90,
    "Bi-Weekly":15,
    "Annually":365,
    "Every 3 Months":90
}

df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)
```

```python
import sqlite3

conn = sqlite3.connect("project.db")
```

```python
df.to_sql("data_table", conn, if_exists="replace", index=False)
```

```python
query = "SELECT * FROM data_table LIMIT 5"
pd.read_sql(query, conn)
```

```python
# 1.What is the total revenue generated by male vs. female customers?
query = "SELECT gender, SUM(purchase_amount) AS revenue FROM data_table GROUP BY gender"
pd.read_sql(query, conn)
```

```python
# 2.Which customers used a discount but still spent more than the average purchase amount?
query = """SELECT customer_id, purchase_amount FROM data_table
WHERE discount_applied = 'Yes'
AND purchase_amount > (
    SELECT AVG(purchase_amount) FROM data_table)"""
pd.read_sql(query, conn)
```

```python
# 3.Which are the top 5 products with the highest average review rating?
query = """SELECT item_purchased, ROUND(AVG(review_rating),2) as "Average Product Rating"
FROM data_table
group by item_purchased
order by AVG(review_rating) desc
LIMIT 5;"""
pd.read_sql(query, conn)
```

```python
# 4.Compare the average Purchase Amounts between Standard and Express Shipping.
query = """SELECT shipping_type, ROUND(AVG(purchase_amount),2) FROM data_table
WHERE shipping_type in ("Standard","Express")
group by shipping_type"""
pd.read_sql(query, conn)
```

```python
# 5.Do subscribed customers spend more? Compare average spend and total revenue between subscribers and non-subscribers.
query = """SELECT subscription_status, COUNT(customer_id) as total_customers,
       ROUND(AVG(purchase_amount),2) AS avg_spend,
       SUM(purchase_amount) AS total_revenue
FROM data_table
GROUP BY subscription_status
order by total_revenue, avg_spend desc;"""
pd.read_sql(query, conn)
```

```python
# 6.Which 5 products have the highest percentage of purchases with discounts applied?
query = """SELECT item_purchased ,ROUND(100*SUM(CASE WHEN discount_applied = 'Yes' Then 1 ELSE 0 END)/COUNT(*),2) as discount_rate
FROM data_table
group by item_purchased
order by discount_rate desc
limit 5;"""
pd.read_sql(query, conn)
```

```python
# 7.Segment customers into New, Returning, and Loyal based on their total number of previous purchases, and show the count of each segment.
query = """SELECT
CASE
    WHEN previous_purchases = 1 THEN 'New'
    WHEN previous_purchases BETWEEN 2 AND 10 THEN 'Returning'
    ELSE 'Loyal'
END AS customer_segment,
COUNT(*) AS total_customers
FROM data_table
GROUP BY customer_segment;"""
pd.read_sql(query, conn)
```

```python
# 8.What are the top 3 most purchased products within each category?
query = """SELECT rank ,category, item_purchased, total_orders
FROM (
    SELECT category, item_purchased, COUNT(*) AS total_orders,
           ROW_NUMBER() OVER (PARTITION BY category ORDER BY COUNT(*) DESC) AS rank
    FROM data_table
    GROUP BY category, item_purchased
)
WHERE rank <= 3;
"""
pd.read_sql(query, conn)
```

```python
# 9.Are customers who are repeat buyers (more than 5 previous purchases) also likely to subscribe?
query = """SELECT subscription_status,
       COUNT(*) AS repeat_buyers
FROM data_table
WHERE previous_purchases > 5
GROUP BY subscription_status;
"""
pd.read_sql(query, conn)
```

```python
# 10.What is the revenue contribution of each age group?
query = """SELECT age_group, SUM(purchase_amount) AS total_revenue
FROM data_table
GROUP BY age_group
ORDER BY total_revenue DESC;
"""
pd.read_sql(query, conn)
```

```python
result=pd.read_sql(query, conn)
result.to_csv("Customer_shopping_FinalClean.csv", index=False)
```
